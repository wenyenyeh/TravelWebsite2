```python
# 安裝 pycaret
!pip install pycaret

# 加載其他必要的庫
import pandas as pd
from pycaret.classification import *
```

```python
# 讀取 Titanic 數據集
data = pd.read_csv('train.csv')

# 顯示數據集的前幾行
data.head()
```

```python
# 設置 PyCaret 環境
clf1 = setup(data=data, target='Survived', session_id=123, verbose=False)
```

```python
# 比較多種分類模型
best_models = compare_models(n_select=16)

# 顯示最佳模型列表
print(best_models)
```

```python
# 訓練最佳模型
final_model = finalize_model(best_models[0])

# 使用最佳模型進行預測
predictions = predict_model(final_model)

# 查看預測結果
predictions.head()
```

```python
# 保存最終模型
save_model(final_model, 'final_titanic_model')

# 提供下載鏈接
from google.colab import files
files.download('final_titanic_model.pkl')
```
